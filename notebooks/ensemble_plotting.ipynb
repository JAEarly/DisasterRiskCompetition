```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
kfold_acc = [0.7459677419354839, 0.8004032258064516, 0.7990591397849462, 0.7997311827956989, 0.771505376344086]
kfold_loss = [0.6568701564857654, 0.5308194591717604, 0.5459091197646108, 0.5107148430416278, 0.5729789380890115]
k = len(kfold_acc)
x = range(1, k+1)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 5))
axes[0].plot(x, kfold_acc)
axes[0].set_xlim(1, k)
axes[0].set_xlabel('k')
axes[0].set_ylabel('Acc')

axes[1].plot(x, kfold_loss)
axes[1].set_xlim(1, k)
axes[1].set_xlabel('k')
axes[1].set_ylabel('Loss')
```
